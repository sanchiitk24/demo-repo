**PORTFOLIO ANALYSIS FUNCTION**

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def analyze_portfolio(array_tickers,array_weightage) :
  portfolio_daily_return = 0
  for ticker,weightage in zip(array_tickers,array_weightage) :
    df_ticker = yf.download(ticker ,period='6mo')
    df_ticker_close = df_ticker['Adj Close']
    daily_returns = df_ticker_close.pct_change()
    weighted_daily_returns = daily_returns*weightage
    portfolio_daily_return = weighted_daily_returns.sum(axis = 1)
    final_value = 100000*np.prod(1+portfolio_daily_return)
  return final_value

def portfolio_return(array_tickers,array_weightage) :
  for ticker,weightage in zip(array_tickers,array_weightage) :
    df_ticker = yf.download(ticker ,period='6mo')
    df_ticker_close = df_ticker['Adj Close']
    daily_returns = df_ticker_close.pct_change()
    weighted_daily_returns = daily_returns*weightage
    portfolio_daily_returns = weighted_daily_returns.sum(axis=1)
  return portfolio_daily_returns
array_weightage_1 = [0.3 , 0.2 , 0.2 , 0.2 ,0.1]
array_weightage_2 = [0.3 , 0.2 , 0.1 , 0.2 ,0.2]
array_tickers = ["AAPL" ,"GOOGL" ,"MSFT" ,"TSLA" ,"NVDA"]
y =(analyze_portfolio(array_tickers,array_weightage_1))
#z = (analyze_portfolio(array_tickers,array_weightage_2))
y_ =(portfolio_return(array_tickers,array_weightage_1))
z_ = (portfolio_return(array_tickers,array_weightage_2))
print(y)

#print(z)

df_1 = yf.download("AAPL" ,period='6mo')
x = df_1.index.tolist()
plt.plot(x,y_)
plt.plot(x,z_)

```
